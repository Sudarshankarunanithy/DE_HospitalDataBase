```python
table_name = []
for i in dbutils.fs.ls('mnt/silver/hospital_healthcare/'):
    table_name.append(i.name.split('/')[0])
```

```python
table_name
```

```python
%python
for name in table_name:
    path = '/mnt/silver/hospital_healthcare/' + name
    print(path)
    df = spark.read.format('delta').load(path)
    output_path = '/mnt/gold/hospital_healthcare/' + name + '/'
    df.write.format('delta').mode('overwrite').save(output_path)
```
